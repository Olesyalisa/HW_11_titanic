```python
import pandas as pd
```

```python
df = pd.read_csv('train (1).csv')
```

```python
df.head(20)
```

```python
df.Survived.value_counts()
```

**Вывод:** \
**_549_** пассажиров - _погибло_ \
**_342_** пассажиров - _выжило_

```python
df.Pclass.value_counts()
```

**Вывод:** \
*первый класс* составлял колличество пассажиров - **_216_** \
*второй класс* составлял колличество пассажиров - **_184_** \
*третий класс* составлял колличество пассажиров - **_491_** 


```python
df.groupby('Pclass', as_index = False) \
    .agg({'Survived': 'sum'}) \
    .sort_values('Pclass', ascending = False)
```

**Вывод:** \
Из общего колличество выживыших: \
*первый класс*, составляет большую часть выживших, а то есть - **136** пассажиров из **216**. \
*Третий класс*, составляет **119** выживших пассажиров из **491**. \
*Второй класс*, составляет **87** выживших пассажиров из **184**.

```python
df.Sex.value_counts()
```

```python
df.groupby('Sex', as_index = False) \
    .agg({'Survived': 'sum'})
```

**Вывод:** 
1. Всего на борту было: **577** *пассажиров мужского пола* и **314** *пассажиров женского пола* \
2. Выжили: 
**_233 пас женского пола_**; (из 314) \
**_109 пас мужского пола_**. (из 577) 

```python
df.groupby('Embarked', as_index = False) \
    .agg({'Survived': 'sum'}) \
    .sort_values('Embarked', ascending = False)
```
